```python
f_seq = open("orf.txt", "r")
x = f_seq.readlines()
#print(x)
```

```python
ol=[]
for i in x:
    i = i[:-1]
    #print(i)
    k=0
    il=[]
    il.append(len(i))
    while k<=len(i):
        codon=i[k:k+3]
        start_codon=0
        flag=True
        dic = []
        #GTA GTT GTC GTG 
        if codon == "GTA" or codon == "GTT" or codon == "GTC" or codon == "GTG" and flag==True:
            print("GT "+ str(k))
            #k=k+3
            dic.append(k)
            start_codon=k
            k=k+3
            flag=False
            flag2=True
            while(flag2 and k<len(i)):
                    #AGA AGT AGC AGG AAG TAG CAG GAG
                    if i[k:k+3] == "AGA" or i[k:k+3] == "AGT" or i[k:k+3] == "AGC" or i[k:k+3] == "AGG" or i[k:k+3] == "AAG" or i[k:k+3] == "TAG" or i[k:k+3] == "CAG" or i[k:k+3] == "GAG":
                        if k-start_codon <= 180:
                            k=k+3
                            continue
                        dic.append(k)
                        print("AG "+ str(k))
                       #genes.append(seq[start_codon:k])
                        flag2=False
                        k=k+3
                    else:
                        k=k+3
            il.append(dic)
        else:
            k=k+3
    ol.append(il)
#print(ol)
print(len(ol))
```

```python
gt_donar=[]
for j in range(0,len(x)):
    seq=x[j]
    seq=seq[:-1]
    l=ol[j]
    for m in range(1,len(l)):
        for z in range(0,5):
            codon=seq[l[m][0]-(30-z):l[m][0]+(30+z)]
            while(len(codon)<60):
                codon=codon+'A'
            #print(len(codon))
            if len(codon) < 60:
                print("BUG")
            gt_donar.append(codon)
#print(gt_donar)
print(len(gt_donar))
```

```python
ag_acceptor=[]
for j in range(0,len(x)):
    seq=x[j]
    seq=seq[:-1]
    l=ol[j]
    for m in range(1,len(l)):
        for z in range(0,5):
            try:
                codon=seq[l[m][1]-(30-z):l[m][1]+(30+z)]
                while(len(codon)<60):
                    codon=codon+'A'
                if len(codon) < 60:
                    print("BUG")
            except:
                while(len(codon)<60):
                    codon=codon+'A'
                if len(codon) < 60:
                    print("BUG")
            ag_acceptor.append(codon)
#print(ag_acceptor)
print(len(ag_acceptor))
```
